```python
import torch
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
mu = torch.rand(2) + 10
# unconstrained, we can do log_sigma_squared.exp() to get variance
# and (0.5 * log_sigma_squared).exp() to get standard deviation,
# which will be both > 0 as expected
log_sigma_squared = torch.rand(2) * 5
```

```python
# expect std and not variance, so it exp(0.5 * log_sigma_squared)
distribution = torch.distributions.Normal(mu, (0.5 * log_sigma_squared).exp())
```

```python
# sampe
distribution.sample()
```

```python
# we can also sample several values at the same time
samples = distribution.sample((200,))
samples.shape
```

```python
plt.scatter(samples[:, 0], samples[:, 1])
```

```python
# we can also sample using the reparameterization trick
# note that we must pass 2D objects
normal_distrib = torch.distributions.Normal(torch.zeros(2), torch.ones(2))
```

```python
normal_distrib.sample()
```

```python
normal_samples = normal_distrib.sample((200,))
reparam_samples = mu + (0.5 * log_sigma_squared).exp() * normal_samples

reparam_samples.shape
```

```python
# we can check that the two sample sets comes from the same distribution

# blue: original distribution samples
# green: sample from N(0, 1)
# orange: samples after reparameterization
plt.scatter(samples[:, 0], samples[:, 1], color="blue")
plt.scatter(reparam_samples[:, 0], reparam_samples[:, 1], color="orange")
plt.scatter(normal_samples[:, 0], normal_samples[:, 1], color="green")
```
