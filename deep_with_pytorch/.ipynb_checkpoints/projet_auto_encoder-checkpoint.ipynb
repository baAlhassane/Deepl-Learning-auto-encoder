```python
decode_modele=Decoder(784,784)
decode_criterion=nn.MSELoss()
decode_optimizer=torch.optim.Adam(decode_modele.parameters(),lr=1e-3,weight_decay=1e-5)
decode_modele
```

## Partie 1-b : Entrainement du modèle

```python
num_epochs=10
outputs=[]
for epoch in range(num_epochs):
    for (img,_) in data_loader:
        img=img.reshape(-1,28*28)
        recon=decode_modele(img)
        loss=decode_criterion(recon,img)#loss entre image autoencoder et image original
        decode_optimizer.zero_grad()
        loss.backward()
        decode_optimizer.step()
    print(f'Epoch:{epoch+1},Loss:{loss.item():.4f}' )
    outputs.append((epoch,img,recon))
        
```

## Part 2: Data generation with an Variational Auto-encoder


```python
class VAEEncoder(nn.Module):
    def __init__(self, dim_input, dim_latent):
        super().__init__()
        # TODO
        self.dim_input=dim_input
        self.dim_latent=dim_latent
        self.encoder=nn.Sequential(
            nn.Linear(dim_input,dim_latent),#reduce size :N,784->N,128
            nn.ReLU(),
            nn.Linear(dim_latent,128),
            nn.Linear(128,64),
            nn.ReLU(),
            nn.Linear(64,12),
            nn.ReLU(),
            nn.Linear(12,dim_latent),#size here is N by 3
            #here we don't need the activation fonction
        )
        self.encoder_mu=nn.Linear(in_features=784,out_features=784)#voir video 1h11min
        self.encoder_sigma=nn.Linear(in_features=784,out_features=784)
        
        self.decoder=nn.Sequential(
            nn.Linear(dim_latent,64),
            nn.ReLU(),
            nn.Linear(64,dim_input),
            nn.Sigmoid(),
            
        )
    def decoder_(self,z) :
        return self.decoder(z)
            
    def forward(self, inputs):
        # TODO
        # mu = mean
        # log_sigma_squared = log varianc
        # The idea is that you use two different output projection:
        # one for the mean, one for the log_sigma_squared
        # but all other layers are shared
        out=self.encoder(inputs)
        out_flat=out.view(inputs.size(0),-1)
        z_sigma=self.encoder_sigma.forward(out_flat)
        z_mu=self.encoder_mu.forward(out_flat)
        eps=torch.normal(mean=0.0,std=1.0,size=z_mu.size())
        z=z_mu+z_sigma*eps
        z_2d=z.view(inputs.size(0),-1,1,1)
        #y_prim=self.decoder(z_2d)
        return z,z_mu,z_sigma
```

##  Partie 2-a  Instanciaton du modele 

```python
VAE_model=VAEEncoder(784,784)
VAE_criterion=nn.MSELoss()
VAE_optimizer=torch.optim.Adam(VAE_model.parameters(),lr=1e-3,weight_decay=1e-5)
VAE_model
```

##  Partie 2-b Fonction loss du modele des auto-encodeur variationnel

```python
def loss_function(inputs, recon,mu, logvar):
    #reconstruction loss
    #mse=VAE_criterion(inputs.view(-1,784),recon)
    mse=nn.MSELoss()
    loss_mse=mse(inputs, recon)
    #logvar=torch.log(var**2)
    kl_loss=0.5*torch.sum(logvar.exp()-logvar-1+mu.pow(2))
    loss=loss_mse+kl_loss
    return loss
```

```python
#training 
recon,mu,var=VAE_model(images)
decode=VAE_model.decoder_(recon)
#logvar=torch.log(logvar.pow(2))
#var=torch.log(var.pow(2))
decode.shape
loss_function(images[0], recon[0],mu[0], var[0])
```

##  Partie 2-c: Traitement des auto-encodeur variationnel  du modele

```python
num_epochs=10
outputs=[]
for i in range(num_epochs):
    for (img,_) in data_loader:
        img=img.reshape(-1,28*28)
        recon,mu,var=VAE_model(img)
        loss=loss_function(img[i], recon[i],mu[i], var[i])#loss entre image autoencoder et image original
        decode_optimizer.zero_grad()
        loss.backward()
        decode_optimizer.step()
    print(f'Epoch:{i+1},Loss:{loss.item():.4f}' )
    outputs.append((i,img,recon))
        
```
